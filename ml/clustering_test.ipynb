```python
def tokenize_and_stem(allwords_tokenized):
    return [stemmer.stem(t) for t in allwords_tokenized]

def tokenize(text):
    text = regex.sub(' ', text).lower().split(' ')
    filtered_text = []
#     global english_ratio
    for token in text:
        if re.search('[a-zA-Z]', token):
            if token not in stopwords and len(token)>2:
                filtered_text.append(token)
    return [el for el in filtered_text if el != '']
#     english_ratio = len(english_text)/len(filtered_text)
#     print("{} / {} = {}".format(len(english_text), len(filtered_text), english_ratio))
#     print(english_text, '\n')
    
#     print(filtered_text)
```

```python
totalvocab_stemmed = []
totalvocab_tokenized = []
titles = []
c = 0
all_texts = []
for root, subdirs, files in os.walk(input_folder_path):
        for file in files:
            c += 1
            if c == 1663: break #  test purposes
            if c % 100 == 0: print("#{}  File:{}".format(c, str(file)))  
                
            file_path = os.path.join(root, file)
            with open(file_path, 'r') as f:
                data = f.read()

              
            allwords_tokenized = tokenize(data)
#             print(allwords_tokenized)
#             print('eng_ratio = {}\n'.format(english_ratio))
            totalvocab_tokenized.extend(allwords_tokenized)
                
            titles.append(file)
            all_texts.append(data)
                
#                 t = time()
            allwords_stemmed = tokenize_and_stem(allwords_tokenized)
#                 print('Time = {}'.format(time()-t))
                
            totalvocab_stemmed.extend(allwords_stemmed)
                
```

```python
num_clusters = 5

svd = TruncatedSVD(num_clusters, )
normalizer = Normalizer(copy=False)
lsa = make_pipeline(svd, normalizer)

tfidf_matrix = lsa.fit_transform(tfidf_matrix)


km = KMeans(n_clusters=num_clusters, init='k-means++', max_iter=100, n_init=1)
km.fit(tfidf_matrix)
```

```python
original_space_centroids = svd.inverse_transform(km.cluster_centers_)
order_centroids = original_space_centroids.argsort()[:, ::-1]
terms = tfidf_vectorizer.get_feature_names()
for i in range(5):
    print("Cluster %d:" % i, end='\n')
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind], end='')
    print()
    print()
```
